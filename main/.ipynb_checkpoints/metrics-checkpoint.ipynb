```python
import pickle
import os
import numpy as np
import matplotlib.pyplot as plt
file = os.path.join('./', 'list_hessians_first.pkl')
with open(file, 'rb') as f_in:
    list_hessians_first = pickle.load(f_in)

file = os.path.join('./', 'list_hessians_last.pkl')
with open(file, 'rb') as f_in:
    list_hessians_last = pickle.load(f_in)
    
file = os.path.join('./', 'list_grads_first.pkl')
with open(file, 'rb') as f_in:
    list_grads_first = pickle.load(f_in)
# list_grads_first = [np.absolute(g) for g in list_grads_first]

file = os.path.join('./', 'list_grads_last.pkl')
with open(file, 'rb') as f_in:
    list_grads_last = pickle.load(f_in)
# list_grads_last = [np.absolute(g) for g in list_grads_last]    
```

```python
num_bins = 2000
n, bins, patches = plt.hist(list_hessians_first, num_bins, 
                            density = 1, 
                            color ='red',
                            alpha = 0.6, label='mid-training, e=0')
n, bins, patches = plt.hist(list_hessians_last, num_bins, 
                            density = 1, 
                            color ='blue',
                            alpha = 0.6, label='post-training')
plt.xlim([-0.0001, 0.0015])
plt.legend()
plt.ylabel('hessian value')

plt.figure()
n, bins, patches = plt.hist(list_grads_first, num_bins, 
                            density = 1, 
                            color ='red',
                            alpha = 0.6, label='mid-training, e=0')
n, bins, patches = plt.hist(list_grads_last, num_bins, 
                            density = 1, 
                            color ='blue',
                            alpha = 0.6, label='post-training')
plt.xlim([-0.003, 0.003])
plt.legend()
plt.ylabel('grad value')
```
