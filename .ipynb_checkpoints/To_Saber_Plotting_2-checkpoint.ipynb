```python
def read_results(alg='FedAvg', seed=0, root='/home/guojun-z/propfair/results/CIFAR10/label-10'):
    log = os.path.join(root, alg, f'seed_{seed}', 'log.json')
    results = []
    with open(log, 'r') as f:
        for line in f:
            results.append(json.loads(line))
    return results[-1]['accs'], results[-1]['losses']
```

```python
accs = read_json(log)[-1]['accs']
```

```python
from utils.stat import mean_std
from utils.io import round_list
```

```python
mean, std = mean_std(accs)
```

```python
root='/home/guojun-z/propfair/results/CIFAR10/label-10'
accs = {}
for alg in algs:
    for seed in range(3):
        log = os.path.join(root, alg, f'seed_{seed}', 'log.json')
        accs[(alg, seed)] = np.sort(read_json(log)[-1]['accs'])
```

```python
means, stds = {}, {}
for alg in algs:
    means[alg] = np.mean([accs[(alg, i)] for i in range(3)], axis=0)
    stds[alg] = np.std([accs[(alg, i)] for i in range(3)], axis=0)
```

```python
def mean_worst(accs):
    return np.mean(accs), np.min(accs)
```

```python
plots_2d = {}
for alg in algs:
    plots_2d[alg] = []
    for seed in range(3):
        mean, worst = mean_worst(accs[(alg, seed)])
        plots_2d[alg].append((mean, worst))
```

```python
def tuple_to_list(list_):
    x_l, y_l = [], []
    for x, y in list_:
        x_l.append(x)
        y_l.append(y)
    return y_l, x_l
```

```python
lists = {}
for alg in algs:
    lists[alg] = tuple_to_list(plots_2d[alg])
```
